```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import yfinance as yf
yf.pdr_override()
from pandas_datareader import data as wb
```

```python
Data_Stock_Prices = yf.download('NFLX', start='2010-01-01', end='2024-04-12')
```
